```python

from bmcs_cross_section.api import MKappa, EC2, ACI440, AnaFRPBending, ReinfLayer
import bmcs_utils.api as bu 
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


b_n = [200, 200, 300, 300]
d_n = [370, 470, 570, 670]

grid_name = ['Q47', 'Q71', 'Q85', 'Q95']
a_nm_n = [116, 173, 210, 232]
f_fu_n = [1250, 1200, 1250, 1200]
E_f_n = [102000, 100000, 102000, 100000]

f_ck_n = 55
c_nom = 30

b_list = []
d_list = []
grid_name_list = []
a_nm_list = []
A_f_list = []
f_fu_list = []
E_f_list = []
M_EC2_list = []

for i in range(0, len(b_n)): # loop through the four defined cross-sections
    for j in range(0, len(a_nm_n)): # loop through the four solidian GRID 
        
        A_f_n = a_nm_n[j]*((b_n[i]-2*c_nom)/1000)
        
        M_EC2 = EC2.get_M_Ed(b=b_n[i], d=d_n[i], As=A_f_n, f_ck=f_ck_n, f_yk=f_fu_n[j])
        print(M_EC2[0]/1e6)
        
        b_list.append(b_n[i])
        d_list.append(d_n[i])
        grid_name_list.append(grid_name[j])
        a_nm_list.append(a_nm_n[j])
        A_f_list.append(A_f_n)
        f_fu_list.append(f_fu_n[j])
        E_f_list.append(E_f_n[j])
        M_EC2_list.append(M_EC2[0]/1e6)


        print(f'b:{b_n[i]} | d:{d_n[i]} | Grid:{grid_name[j]} | a_nm:{a_nm_n[j]} | A_f_n:{A_f_n:.2f} | f_fu_n:{f_fu_n[j]} | E_f_n:{E_f_n[j]} | M_EC2 :{M_EC2[0]/1e6:.2f}')

M_solidian_list = [6.73, 9.6, 12.11, 12.84, 8.57, 12.23, 15.43, 16.35, 15.61, 22.29, 28.13, 29.81, 18.37, 26.23, 33.11, 35.09 ]


data = {
    'b' : b_list, 
    'd' : d_list, 
    'Grid' : grid_name_list, 
    'a_nm' : a_nm_list, 
    'A_f' : A_f_list, 
    'f_fu' : f_fu_list, 
    'E_f_n' : E_f_list, 
    'M_EC2' : M_EC2_list,
    'M_solidian' : M_solidian_list

}
df = pd.DataFrame(data)
```
